```python
import pandas as pd
import json
import requests
```

```python
df = pd.read_json('query_week')
df
```

```python
df1 = pd.DataFrame.from_records(df['entries'])
df1
```

```python
#笨方法

cols = list(i for i in df['entries'][0].keys())
lists = []
for i in cols:
    i = []
    lists.append(i)
    
for i in df['entries']:
    for j,k in zip(cols,lists):
        k.append(i[j])

df1 = pd.DataFrame(columns=cols)
for i,j in zip(cols,lists):
    df1[i]=j
df1
```

```python
df1.to_excel('空氣盒子.xlsx',columns=cols,encoding='utf-8')
```
