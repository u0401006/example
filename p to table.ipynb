```python
import re
import pandas as pd

#讀取來源檔
with open('hsin.txt','r') as IN:
    text = IN.readlines()

#確認欄位，並判斷每筆資料中有沒有相同欄位
cols=['筆數','姓名：','字號：','日文名：','生日：','籍貫：','住址：','家庭：','學歷：','資產：','經歷：','出處：','備註：']
for i in range(88):
    for j in cols:
        if j in text[i]:
            pass
        else:
            text[i] = text[i]+'\t'+j

#把每一列資料分割成list，以利迴圈檢查
text1=[]
for i in text:
    a = i.split('\t')
    text1.append(a)
    
#製作每個欄位的容器，方便之後倒入table
index=[]
name=[]
nname=[]
jname=[]
bir=[]
badd=[]
add=[]
fam=[]
edu=[]
wor=[]
wel=[]
ref=[]
note=[]
lists=[index,name,nname,jname,bir,badd,add,fam,edu,wel,wor,ref,note]

#檢查內容並倒入欄位容器
for i in text1:
    for j in i:
        for k in range(13):
            if cols[k] in j:
                lists[k].append(j.replace(cols[k],''))

#製作dataframe
df =pd.DataFrame(columns=cols)
for i,j in zip(cols,lists):
    df[i]=j
df.to_csv('hsin.csv',encoding='utf-8')
```
